```python
import pandas as pd
```

```python
df = pd.read_csv('data/acled_social_iso3.csv')
```

```python
month_range = list(range(len(df["EVENT_DATE_MONTH"].unique())))

month_dict = {month:month_range[i] for i,month in enumerate(df["EVENT_DATE_MONTH"].unique())}
```

```python
def apply_month(row):
    row_d = row.to_dict()
    row_d["month_int"] = month_dict[row_d['EVENT_DATE_MONTH']]
    
    return pd.Series(row_d)
    
```

```python
df= df.apply(apply_month, axis=1)
```

```python
import geopandas as gpd

# Access built-in Natural Earth data via GeoPandas
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Get a list (dataframe) of country centroids
centroids = world.centroid
```

```python
point = world[world['iso_a3'] == 'DZA'].centroid
```

```python
latitude = point.to_list()[0].xy[1].tolist()[0]
longitude = point.to_list()[0].xy[0].tolist()[0]
```

```python
def apply_coords(row):
    row_d = row.to_dict()
    point = world[world['iso_a3'] == row_d['iso3']].centroid
    row_d["lat"] = point.to_list()[0].xy[1].tolist()[0]
    row_d["lon"] = point.to_list()[0].xy[0].tolist()[0]
    return pd.Series(row_d)
    
```

```python
df = df.apply(apply_coords, axis=1)
```
