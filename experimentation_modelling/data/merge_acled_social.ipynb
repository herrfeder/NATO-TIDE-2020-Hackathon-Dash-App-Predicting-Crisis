```python
import pandas as pd
```

```python
df_social = pd.read_csv("social_data.csv")
```

```python
df_social['year'] = pd.to_datetime(df_social['Date']).dt.year
```

```python
df_social
```

```python
df_acled = pd.read_csv('acled_africa_month_comp.csv')
```

```python
df_acled
```

```python
rel_cat = ["Life expectancy at birth, total (years)",
           "GDP (current US$)",
           "Death rate, crude (per 1,000 people)",
           "Population growth (annual %)",
           "Urban population (% of total)",
           "Central government, Fiscal Balance (Current US $)",
           "Population, Total"
          ]
```

```python
cat_short = ["life_exp",
           "gdp",
           "death_rate",
           "pop_growth",
           "urb_pop",
           "gov_fin",
           "pop_total"
          ]
```

```python
for cou in df_acled.COUNTRY.unique():
    for year in df_acled.YEAR.unique():
        for cat in rel_cat:
            try:
                value = df_social[(df_social["IndicatorName"] == cat) & (df_social["CountryName"] == cou) & (df_social['year'] == year)]['Value'].values[0]
            except:
                value=0
            indexes = df_acled[(df_acled["YEAR"] == year) & (df_acled["COUNTRY"] == cou)].index
            df_acled.loc[indexes, cat] = value
```

```python
df_acled.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
df_acled.to_csv('acled_social.csv')
```
