```python
df_time = df[["EVENT_DATE_MONTH", "Battles"]]
```

```python
df_time["EVENT_DATE_MONTH"] = pd.to_datetime(df_time["EVENT_DATE_MONTH"])
```

```python
df["Violence against civilians"].value_counts()
```

```python
df.head()
```

```python
def fatalities(row):
    if row["FATALITIES"] > 50:
        return 1
    else:
        return 0
    
def violence_explosion(row):
    if row["Explosions/Remote violence"] > 10:
        return 1
    else:
        return 0

def protests(row):
    if row["Protests"] > 10:
        return 1
    else:
        return 0
    
def violence_civilians(row):
    if row["Violence against civilians"] > 10:
        return 1
    else:
        return 0
    
    
```

```python
df_time["fat_reg"] = df.apply(fatalities, axis=1)
df_time["exp_reg"] = df.apply(violence_explosion, axis=1)
df_time["pro_reg"] = df.apply(protests, axis=1)
df_time["civ_reg"] = df.apply(violence_civilians, axis=1)
```

```python
from fbprophet import Prophet
```

```python
df_time
```

```python
df_time.rename(columns={"EVENT_DATE_MONTH":"ds", "Battles":"y"}, inplace=True)
```

```python
m = Prophet(daily_seasonality=True, weekly_seasonality=True)
m.add_regressor('fat_reg')
m.add_regressor('exp_reg')
m.add_regressor('pro_reg')
m.add_regressor('civ_reg')

m.fit(df_time, )
future = m.make_future_dataframe(periods=df_time.shape[0])

future["fat_reg"] = df.apply(fatalities, axis=1)
future["exp_reg"] = df.apply(violence_explosion, axis=1)
future["pro_reg"] = df.apply(protests, axis=1)
future["civ_reg"] = df.apply(violence_civilians, axis=1)

forecast = m.predict(future)
m.plot_components(forecast)
```
